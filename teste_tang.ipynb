```python
import tanglegram as tg
import matplotlib.pyplot as plt 
import pandas as pd
```

```python
labelsA=['A', 'B', 'C', 'D']
```

```python
labelsB=['B', 'A', 'C', 'D']
```

```python
data= [[0, .1, .4, .3],
       [.1, 0, .5, .6],
       [.4, .5, 0, .2],
       [.3, .6, .2, 0]]
```

```python
mat1= pd.DataFrame(data, columns=labelsA, index=labelsA)
mat2= pd.DataFrame(data, columns=labelsB, index=labelsB) 
```

```python
fig=tg.plot(mat1, mat2, sort=False)
plt.show()
```
