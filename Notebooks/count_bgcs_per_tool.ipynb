```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Caminho para o arquivo 
file_path = "/home/barbara/documents/RegPatternSearch/contagemtdsBGCS_prismeAS.cvs" 
```

```python
# Lendo o CSV
df = pd.read_csv(file_path, sep=",")
df.set_index('species', inplace=True)
df = df.fillna(0)

# Transpondo o DataFrame para ter espécies no eixo x
df_t = df.T  # agora as colunas são espécies, linhas são categorias
```

```python
# Criando o heatmap
plt.figure(figsize=(30,30))
sns.heatmap(df_t, 
            annot=False,       # se quiser valores, use True
            fmt=".0f", 
            cmap="viridis", 
            linewidths=2.0,
            cbar_kws={'label': 'Contagem'})

# Títulos e rótulos
plt.title("Distribuição de BGCs: antiSMASH vs PRISM")
plt.xlabel("Espécies")
plt.ylabel("Categorias")
```

```python
plt.tight_layout()
plt.show()
```
